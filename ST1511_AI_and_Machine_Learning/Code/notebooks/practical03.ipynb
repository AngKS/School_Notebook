##Check for outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns

vals = wineDF.columns.values
colNum = len(wineDF.columns)
rowNum = (len(vals) -1 ) / colNum

plt.figure(figsize=(colNum, (5*rowNum)))
for i in range(0, len(vals)):
    plt.subplot(rowNum + 1, colNum, i+1)
    sns.set_style('whitegrid')
    sns.boxplot(data=wineDF[vals[i]], color="blue", orient="Vertical")
    plt.tight_layout()
```

```python
fig, ax = plt.subplots(1, 12, figsize=(16, 5))
col = wineDF.columns

col = wineDF.columns
for i in range(12):
    wineDF[col[i]].plot(kind = "box", ax=ax[i])
plt.tight_layout()
plt.show()
```

```python
# To check for distribution skewness

plt.figure(figsize=(2*(colNum), 5*rowNum))

for i in range(0, len(wineDF.columns.values)):
    plt.subplot(rowNum + 1, colNum, i+1)
    sns.displot(x=wineDF[vals[i]], kde=True)
```

```python
# Feature Engineering

wineDF["Strong"] = wineDF['alcohol'] >= 12.5
wineDF.head()
```

```python
sns.countplot(x='quality', hue="Strong", data=wineDF)
# Stronger alcohol tends to be of higher quality as well.
```

```python
# Relationship between different quality and strength of alcohol

sns.pointplot(x="quality", y="Strong", data=wineDF)
```

```python
plt.figure(figsize=(6, 4))

sns.heatmap(wineDF.corr(), cmap="Blues", annot=False)
```

```python
# Quality Correlation Matrix

k = 5
cols = wineDF.corr().nlargest(k, 'quality')['quality'].index
cm = wineDF[cols].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(cm, annot=True, cmap="viridis")
```
